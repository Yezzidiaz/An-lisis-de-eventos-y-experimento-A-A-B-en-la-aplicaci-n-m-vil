```python
import pandas as pd

# 1. Leer el archivo con detecci√≥n autom√°tica del delimitador
df = pd.read_csv('/datasets/logs_exp_us.csv', sep=None, engine="python")

# 2. Limpiar y renombrar columnas
df.columns = df.columns.str.strip()  # elimina espacios invisibles
df = df.rename(columns={
    'EventName': 'event',
    'DeviceIDHash': 'user_id',
    'EventTimestamp': 'timestamp',
    'ExpId': 'experiment'
})

# 3. Convertir columna de fechas
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 4. Resumen inicial
print("üìä Columnas finales:", df.columns.tolist())
print("\nüîé Vista previa:")
print(df.head())

print("\nCantidad de eventos totales:", len(df))
print("Usuarios √∫nicos:", df['user_id'].nunique())
print("Rango de fechas:", df['timestamp'].min(), "‚Üí", df['timestamp'].max())

print("\nDistribuci√≥n de eventos:")
print(df['event'].value_counts())

print("\nDistribuci√≥n de usuarios por experimento:")
print(df['experiment'].value_counts())
```

**4. Utiliza el embudo de eventos para encontrar la proporci√≥n de usuarios que pasan de una etapa a la siguiente. Por ejemplo, para la secuencia de eventos A ‚Üí B ‚Üí C, calcula la proporci√≥n de usuarios en la etapa B a la cantidad de usuarios en la etapa A y la proporci√≥n de usuarios en la etapa C a la cantidad en la etapa B.**

```python
# Definir el orden del embudo
embudo = ["product_page", "cart_add", "purchase"]

# Diccionario para guardar usuarios √∫nicos en cada paso
usuarios_por_etapa = {}

for etapa in embudo:
    usuarios_por_etapa[etapa] = df.loc[df["event"] == etapa, "user_id"].nunique()

# Calcular proporciones entre etapas consecutivas
conversiones = {}
for i in range(1, len(embudo)):
    etapa_anterior = embudo[i-1]
    etapa_actual = embudo[i]

    usuarios_prev = usuarios_por_etapa[etapa_anterior]
    usuarios_act = usuarios_por_etapa[etapa_actual]

    if usuarios_prev > 0:
        conversiones[f"{etapa_anterior} ‚Üí {etapa_actual}"] = usuarios_act / usuarios_prev
    else:
        conversiones[f"{etapa_anterior} ‚Üí {etapa_actual}"] = 0  # o None si prefieres

# Mostrar resultados
print("Usuarios √∫nicos por etapa:")
for etapa, usuarios in usuarios_por_etapa.items():
    print(f"- {etapa}: {usuarios}")

print("\nTasas de conversi√≥n entre etapas:")
for paso, tasa in conversiones.items():
    print(f"- {paso}: {tasa:.2%}")
```

**5. ¬øEn qu√© etapa pierdes m√°s usuarios?**

```python
# Definir el embudo
embudo = ["product_page", "cart_add", "purchase"]

# Contar usuarios √∫nicos por etapa
usuarios_por_etapa = {}
for etapa in embudo:
    usuarios_por_etapa[etapa] = df[df["event"] == etapa]["user_id"].nunique()

# Calcular tasas de conversi√≥n y p√©rdidas
conversiones = {}
perdidas = {}
for i in range(1, len(embudo)):
    etapa_anterior = embudo[i-1]
    etapa_actual = embudo[i]
    usuarios_prev = usuarios_por_etapa[etapa_anterior]
    usuarios_curr = usuarios_por_etapa[etapa_actual]
    
    if usuarios_prev > 0:
        tasa = usuarios_curr / usuarios_prev
        conversiones[f"{etapa_anterior} ‚Üí {etapa_actual}"] = tasa
        perdidas[f"{etapa_anterior} ‚Üí {etapa_actual}"] = 1 - tasa
    else:
        conversiones[f"{etapa_anterior} ‚Üí {etapa_actual}"] = None
        perdidas[f"{etapa_anterior} ‚Üí {etapa_actual}"] = None

# Mostrar resultados
print("Usuarios √∫nicos por etapa:")
for etapa, n in usuarios_por_etapa.items():
    print(f"- {etapa}: {n}")

print("\nTasas de conversi√≥n:")
for trans, tasa in conversiones.items():
    if tasa is not None:
        print(f"- {trans}: {tasa:.2%}")
    else:
        print(f"- {trans}: no disponible (etapa anterior vac√≠a)")

# Encontrar la etapa con mayor p√©rdida
etapa_mayor_perdida = max(perdidas, key=lambda k: perdidas[k] if perdidas[k] is not None else -1)
print(f"\nüìâ La mayor p√©rdida de usuarios ocurre en: {etapa_mayor_perdida}")
```

**2. Tenemos dos grupos de control en el test A/A, donde comprobamos nuestros mecanismos y c√°lculos. Observa si hay una diferencia estad√≠sticamente significativa entre las muestras 246 y 247.**

```python
from statsmodels.stats.proportion import proportions_ztest

# Usuarios que llegaron al pago por grupo
usuarios_pago = df[df["event"] == "PaymentScreenSuccessful"].groupby("experiment")["user_id"].nunique()

# Usuarios totales por grupo
usuarios_totales = df.groupby("experiment")["user_id"].nunique()

# Extraer SOLO los dos grupos de control usando loc
success = [usuarios_pago.loc[246], usuarios_pago.loc[247]]
nobs = [usuarios_totales.loc[246], usuarios_totales.loc[247]]

# Test de hip√≥tesis
stat, pval = proportions_ztest(success, nobs)

print("Estad√≠stico Z:", stat)
print("p-valor:", pval)
```

Interpretaci√≥n

p-valor < 0.05 ‚Üí diferencias significativas entre los grupos de control (algo raro).

p-valor ‚â• 0.05 ‚Üí no hay diferencias, los controles son equivalentes (lo esperado en A/A).

```python

import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

def comparar_eventos(df, evento, grupo_col="experiment", user_col="user_id", event_col="event"):
    df_evento = df[df[event_col] == evento]
    usuarios_evento = df_evento.groupby(grupo_col)[user_col].nunique()
    total_usuarios = df.groupby(grupo_col)[user_col].nunique()

    grupos = sorted(df[grupo_col].unique())
    if len(grupos) < 2:
        return None

    g1, g2 = grupos[:2]  # tomar los dos grupos de control: 246 y 247
    x = [usuarios_evento.get(g1, 0), usuarios_evento.get(g2, 0)]
    n = [total_usuarios.get(g1, 0), total_usuarios.get(g2, 0)]

    prop_g1 = x[0] / n[0] if n[0] > 0 else 0
    prop_g2 = x[1] / n[1] if n[1] > 0 else 0

    # Casos donde no se puede calcular z-test (todo 0 o todo 100%)
    if (x[0] == 0 and x[1] == 0) or (x[0] == n[0] and x[1] == n[1]):
        return {
            "evento": evento,
            f"usuarios_{g1}": x[0],
            f"total_{g1}": n[0],
            f"proporcion_{g1}": prop_g1,
            f"usuarios_{g2}": x[1],
            f"total_{g2}": n[1],
            f"proporcion_{g2}": prop_g2,
            "z": None,
            "p_value": None,
            "resultado": "No calculable (proporciones extremas)"
        }

    stat, pval = proportions_ztest(count=x, nobs=n)

    return {
        "evento": evento,
        f"usuarios_{g1}": x[0],
        f"total_{g1}": n[0],
        f"proporcion_{g1}": prop_g1,
        f"usuarios_{g2}": x[1],
        f"total_{g2}": n[1],
        f"proporcion_{g2}": prop_g2,
        "z": stat,
        "p_value": pval,
        "resultado": "Significativa" if pval < 0.05 else "No significativa"
    }

def analizar_todos_eventos(df, grupo_col="experiment", user_col="user_id", event_col="event"):
    resultados = []
    for evento in df[event_col].unique():
        res = comparar_eventos(df, evento, grupo_col, user_col, event_col)
        if res:
            resultados.append(res)
    return pd.DataFrame(resultados)

# üëâ Ejecutar
resultado_eventos = analizar_todos_eventos(df)
print(resultado_eventos)
```

**4. Haz lo mismo para el grupo con fuentes alteradas. Compara los resultados con los de cada uno de los grupos de control para cada evento de forma aislada. Compara los resultados con los resultados combinados de los grupos de control. ¬øQu√© conclusiones puedes sacar del experimento?**

```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

def comparar_eventos_pares(df, evento, grupo_a, grupo_b, grupo_col="experiment", user_col="user_id", event_col="event"):
    df_evento = df[df[event_col] == evento]
    usuarios_evento = df_evento.groupby(grupo_col)[user_col].nunique()
    total_usuarios = df.groupby(grupo_col)[user_col].nunique()

    x = [usuarios_evento.get(grupo_a, 0), usuarios_evento.get(grupo_b, 0)]
    n = [total_usuarios.get(grupo_a, 0), total_usuarios.get(grupo_b, 0)]

    prop_a = x[0] / n[0] if n[0] > 0 else 0
    prop_b = x[1] / n[1] if n[1] > 0 else 0

    # Si no se puede calcular estad√≠stica
    if (x[0] == 0 and x[1] == 0) or (x[0] == n[0] and x[1] == n[1]):
        return {
            "evento": evento,
            "grupo_a": grupo_a, "usuarios_a": x[0], "total_a": n[0], "prop_a": prop_a,
            "grupo_b": grupo_b, "usuarios_b": x[1], "total_b": n[1], "prop_b": prop_b,
            "z": None, "p_value": None, "resultado": "No calculable"
        }

    stat, pval = proportions_ztest(count=x, nobs=n)

    return {
        "evento": evento,
        "grupo_a": grupo_a, "usuarios_a": x[0], "total_a": n[0], "prop_a": prop_a,
        "grupo_b": grupo_b, "usuarios_b": x[1], "total_b": n[1], "prop_b": prop_b,
        "z": stat, "p_value": pval,
        "resultado": "Significativa" if pval < 0.05 else "No significativa"
    }

def comparar_experimento(df, grupo_exp=248, controles=[246,247], grupo_col="experiment", user_col="user_id", event_col="event"):
    resultados = []

    # Comparar exp contra cada control individual
    for control in controles:
        for evento in df[event_col].unique():
            res = comparar_eventos_pares(df, evento, grupo_exp, control, grupo_col, user_col, event_col)
            resultados.append(res)

    # Comparar exp contra los controles combinados
    df_controles = df[df[grupo_col].isin(controles)].copy()
    df_controles[grupo_col] = "controles"
    df_exp = df[df[grupo_col] == grupo_exp].copy()
    df_exp[grupo_col] = "experimento"

    df_combinado = pd.concat([df_controles, df_exp])

    for evento in df[event_col].unique():
        res = comparar_eventos_pares(df_combinado, evento, "experimento", "controles", grupo_col, user_col, event_col)
        resultados.append(res)

    return pd.DataFrame(resultados)

# üëâ Ejecutar comparaci√≥n completa
resultado_exp = comparar_experimento(df)
print(resultado_exp)
```

**5. ¬øQu√© nivel de significancia has establecido para probar las hip√≥tesis estad√≠sticas mencionadas anteriormente? Calcula cu√°ntas pruebas de hip√≥tesis estad√≠sticas has realizado. Con un nivel de significancia estad√≠stica de 0.1, uno de cada 10 resultados podr√≠a ser falso. ¬øCu√°l deber√≠a ser el nivel de significancia? Si deseas cambiarlo, vuelve a ejecutar los pasos anteriores y comprueba tus conclusiones.**

```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
from statsmodels.stats.multitest import multipletests

# --- Helper: test proporciones entre dos grupos para un evento ---
def test_event_pair(df, evento, grupo_a, grupo_b, grupo_col="experiment", user_col="user_id", event_col="event"):
    df_evento = df[df[event_col] == evento]
    usuarios_evento = df_evento.groupby(grupo_col)[user_col].nunique()
    total_usuarios = df.groupby(grupo_col)[user_col].nunique()
    x = [usuarios_evento.get(grupo_a, 0), usuarios_evento.get(grupo_b, 0)]
    n = [total_usuarios.get(grupo_a, 0), total_usuarios.get(grupo_b, 0)]
    # si no hay base para el test, devolvemos p=nan
    if n[0] == 0 or n[1] == 0:
        return {"evento": evento, "grupo_a": grupo_a, "grupo_b": grupo_b,
                "x_a": x[0], "n_a": n[0], "x_b": x[1], "n_b": n[1],
                "prop_a": None, "prop_b": None, "z": None, "p": float("nan")}
    # casos extremos donde no tiene sentido z-test
    if (x[0] == 0 and x[1] == 0) or (x[0] == n[0] and x[1] == n[1]):
        prop_a = x[0] / n[0] if n[0]>0 else None
        prop_b = x[1] / n[1] if n[1]>0 else None
        return {"evento": evento, "grupo_a": grupo_a, "grupo_b": grupo_b,
                "x_a": x[0], "n_a": n[0], "x_b": x[1], "n_b": n[1],
                "prop_a": prop_a, "prop_b": prop_b, "z": None, "p": float("nan")}
    stat, pval = proportions_ztest(count=x, nobs=n)
    prop_a = x[0] / n[0]
    prop_b = x[1] / n[1]
    return {"evento": evento, "grupo_a": grupo_a, "grupo_b": grupo_b,
            "x_a": x[0], "n_a": n[0], "x_b": x[1], "n_b": n[1],
            "prop_a": prop_a, "prop_b": prop_b, "z": stat, "p": pval}

# --- Generar todas las pruebas: 248 vs 246, 248 vs 247, 248 vs (246+247) ---
def all_tests(df, exp_group=248, controls=[246,247], event_col="event"):
    resultados = []
    eventos = df[event_col].unique()
    # individual comparisons
    for control in controls:
        for ev in eventos:
            resultados.append(test_event_pair(df, ev, exp_group, control))
    # combined controls: create temporary combined counts
    df_ctrl = df[df["experiment"].isin(controls)].copy()
    df_ctrl["experiment"] = "controls_combined"
    df_exp = df[df["experiment"] == exp_group].copy()
    df_exp["experiment"] = "experiment_248"
    df_comb = pd.concat([df_ctrl, df_exp], ignore_index=True)
    for ev in eventos:
        resultados.append(test_event_pair(df_comb, ev, "experiment_248", "controls_combined", grupo_col="experiment", event_col=event_col))
    return pd.DataFrame(resultados)

# Ejecutar pruebas
res_df = all_tests(df)

# Contar pruebas v√°lidas (con p no-NaN)
m_total = res_df.shape[0]
m_valid = res_df['p'].notna().sum()
print(f"Total tests planned = {m_total}, valid tests with p-values = {m_valid}")

# Aplicar correcci√≥n Bonferroni y BH (FDR)
# Tomamos solo las pruebas con p-values no NaN para la correcci√≥n
pvals = res_df['p'].fillna(1.0).values  # NaN -> 1 (no signific)
# Bonferroni (alpha=0.05 family-wise)
rej_bonf, pvals_bonf, _, _ = multipletests(pvals, alpha=0.05, method='bonferroni')
# Benjamini-Hochberg FDR (q=0.05)
rej_bh, pvals_bh, _, _ = multipletests(pvals, alpha=0.05, method='fdr_bh')

res_df['p_raw'] = res_df['p']
res_df['p_bonf'] = pvals_bonf
res_df['signif_bonf'] = rej_bonf
res_df['p_bh_adj'] = pvals_bh
res_df['signif_bh'] = rej_bh

# Mostrar resumen: cu√°ntas significativas seg√∫n cada correcci√≥n
print("\nSignificativas (raw p < 0.05):", (res_df['p_raw'] < 0.05).sum())
print("Significativas after Bonferroni:", res_df['signif_bonf'].sum())
print("Significativas after BH (FDR 5%):", res_df['signif_bh'].sum())

# Mostrar tabla final
display(res_df.sort_values('p_raw').reset_index(drop=True))
```

C√≥mo interpretar los resultados tras correcci√≥n

Si ninguna prueba sobrevive a Bonferroni - el efecto es fr√°gil; probablemente no hay evidencia fuerte de que las fuentes cambien el comportamiento.

Si alguna prueba sobrevive a Bonferroni - es una se√±al robusta (muy poco probable que sea falso positivo).

Si varias pruebas son significativas √∫nicamente con p < 0.05 crudo pero no con BH ‚Üí hay indicios pero podr√≠amos estar viendo falsos positivos por m√∫ltiples comparaciones. BH te dice la lista de descubrimientos con FDR controlada (√∫til cuando esperas varios efectos).

Recomendaci√≥n final

Practica recomendada para product analytics: usar FDR (Benjamini‚ÄìHochberg) al 5% y reportar los p-values crudos, p-values ajustados y el tama√±o del efecto (diferencia de proporciones).

Si la decisi√≥n es muy costosa/riesgosa (por ejemplo, cambiar UI para todos los usuarios), a√±ade una segunda salvaguarda: Bonferroni Œ±=0.05/15 ‚âà 0.0033 para confirmar solo efectos muy robustos.
